```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Getting the Data

```python
FILE_PATH = os.path.join(os.getcwd(),'Assignment#1_Performance_track_records.csv')

data = pd.read_csv(FILE_PATH)
```

```python
daily_return = data.drop(columns=['Asset', 'DateTime', 'Price_Close'])
daily_return['DateTime'] = pd.to_datetime(data['DateTime'])

daily_return = daily_return.set_index('DateTime')
```

```python
daily_return.head()
```

# Performance Analysis

## Cumulative Return

```python
cum_return = (daily_return + 1).cumprod()
```

__The Total Returns of all strategies are listed below.__

```python
total_return = (daily_return + 1).prod() - 1
total_return.sort_values(ascending=False)
```

```python
cum_return.plot(figsize=(15,12), title='Net Value Path of All Strategies')
```

## Annualized Return

__The Annualized Returns of all strategies are listed below.__

```python
annualized_return = pow(total_return+1, 1/3) - 1
display(annualized_return.sort_values(ascending=False))
```

## Annualized Volatility

__The Annualized Volatility of all strategies are listed below.__

```python
annualized_vol = daily_return.std() * np.sqrt(250)
display(annualized_vol.sort_values(ascending=False))
```

## Sharpe Ratio

__We use 10 year treasury rate as the risk free rate to calculate their Sharpe Ratios.__  
Source: https://www.multpl.com/10-year-treasury-rate/table/by-year

```python
rf_2014 = 0.0286
rf_2015 = 0.0188
rf_2016 = 0.0209
```

```python
rf = pow((1+rf_2014)*(1+rf_2015)*(1+rf_2016), 1/3) - 1
```

__The Sharpe Ratios of all strategies are listed below.__

```python
sharpe_ratio = (annualized_return - rf) / annualized_vol
display(sharpe_ratio.sort_values(ascending=False))
```

## Information Ratio

__We use S&P 500 yearly return as the benchmark to calculate Information Ratio__  
Source: https://ycharts.com/indicators/sandp_500_total_return_annual

```python
bm_2014 = 0.1369
bm_2015 = 0.0138
bm_2016 = 0.1196
```

```python
benchmark = pow((1+bm_2014)*(1+bm_2015)*(1+bm_2016), 1/3) - 1
```

__The Information Ratios of all strategies are listed below.__

```python
information_ratio = (annualized_return - benchmark) / annualized_vol
display(information_ratio.sort_values(ascending=False))
```

## Maximum Drawdown

```python
max_dd = ((cum_return.cummax() - cum_return) / cum_return.cummax()).max()
display(max_dd.sort_values(ascending=False))
```

```python
df = pd.concat([total_return, annualized_return, annualized_vol, 
           sharpe_ratio, information_ratio, max_dd], axis=1)
df.columns = ['Total Return', 'Annualized Return', 'Annualized Volatility', 
                  'Sharpe Ratio', 'Information Ratio', 'Max Drawdown']
```

```python
display(df)
```
