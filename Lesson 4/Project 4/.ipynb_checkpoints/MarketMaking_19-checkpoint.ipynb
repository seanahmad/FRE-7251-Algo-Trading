# Strategies

We build two different strategies to do the trading simulation.
## Strategy 1: Market-Follower
In this strategy, we quote exactly the same as the market price observed, but we would have to be X milliseconds later than the market. For instance, when we observed the market bid price is 1.3554 and market ask price is 1.3562, we would send out the same bid and ask price, but our price is delayed X milliseconds. In the meantime, because our price is the same as the market's, we may not be able to execute the order (probability is 0.5). As is stated, this strategy is a passive management strategy. We do not intervene with how the price is quoted.


## Strategy 2: Position-Watcher
In this strategy, we would adjust our price quote according to our net position and the market price. We want to achieve a relatively lower net position in the whole time window, by actively manage our position with price quote. We would discuss different situation and quote below.
1. When our net open position is between -3 and 3, the net open position is low, and we aggressively want to take more orders. Thus, we would give out a 20% of the spread lower bid price than market bid price and a 20% of the spread higher ask price than the market ask price.
2. When our net open position is between -7 and -3 or between 3 and 7, the net open position is in a medium level, and we would stand neutrally in the market, i.e. follow the market price to give out our price. In this situation, we act the same as the first strategy.
3. When our net open position is less than -7, we have taken a lot of buy orders. We would stay in the bid price market, by increasing 20% of the spread than the market bid price, in order to take more sell orders to offset our negative positon. In the meantime, we would stay out of the ask market, by increasing 20% of the spread than the market ask price, in order to reduce the chance of taking more buy orders.
4. When our net open position is less than -7, we have taken a lot of sell orders. We would stay in the ask price market, by lowering 20% of the spread than the market ask price, in order to take more buy orders to offset our negative positon. In the meantime, we would stay out of the bid market, by lowering 20% of the spread than the market bid price, in order to reduce the chance of taking more sell orders.


## Strategy 3: Trend-Follower
In this strategy, we define two variables called `MA_5` and `MA_20` to monitor the trend of the market price. MA_5 is the the moving average of the past 5 ticks of bid/ask prices. And MA_20 is the moving average of the pst 20 ticks of bid/ask prices. Each time a new tick comes in and we will act and quote differently by comparing the new MA_5 and MA_20.
1. When the MA_5 is larger than the MA_20, we would think that the market is on a trend going up. At this moment, we would aggress the bid market, and retreat from the ask market. We would set our bid price at the market bid price plus 20% of the spread, and set our ask price at the market ask price plus 20% of the spread.
2. When the MA_5 is lower than the MA_20, we would think that the market is on a trend going down. At this moment,  we would aggress the ask market, and retreat from the bid market. We would set our ask price at the market ask price minus 20% of the spread, and set our bid price at the market bid price minus 20% of the spread.

- On 2018-01-07 and 2018-01-08, the Trend-Follower strategy beats the Market-Follower and Position-Watcher in PnL  On these two days, the latter two strategies has negative total PnLs, while Trend-Follower earns positive PnLs.  

- But Trend-Follower does not do a good job all the time. It fails on 2018-01-10. It beats Position-Watcher, but falls behind the simple Market-Follower. If we look at the market price on that day, we will find that the market is on a fast up-trend, but our Trend-Follower does not regonize the trend. According to our expectation, it should regonize the trend, aggres bid market, retreat from ask market, and try to take more sell orders. However, it acts oppositely. When the market is going up, a lot of buy orders are coming in. The Position-Watcher takes all the orders, because at this time there is no inventory in his position, he would aggress the market according to our settings. And in the meantime, the Trend-Follower fails to recognize the up-trend. It also takes in all the sell orders. As the market price goes up, they both take a large loss. The Market-Follower follows the market price to quote. It only has 50% possibility to take in the sell orders, so it takes a less loss.  

- On 2018-01-12, Position-Watcher perform worst. According to our settings, it only cares about the net open position. When net open position is low, it will aggress the market. When net open position is high, it will retreat from the market, and it will not care about whether these aggression or retreat will make profit or not. Position is its first priority. In most of the time, it fails the simple Market-Follower.

## Net Open Position

```python
fig_netpos
```

Compared with the PnL plots in the last section, these net open position plots would give us a better understanding of why certain strategies fail or success.  

- On the first day, Trend-Follower's net open postion goes up, while the other two's positions are going down. And as we stated in the last section, the first day's market is on a up-trend. Those who has a negative postion, taking too much buy orders, would have a terrible loss. And Trend-Follower recognize this up-trend, and aggress the market, thus it makes a positive PnL.  

- Similarly, on the second, Trend-Follower's net open position goes deep down, while the other two's are going up. When the orders are coming in, the market is on a down side. Trend-Follower also recognize this trend and retreat from the bid market, aggress the ask market. It takes lots of buy orders, while the other two are taking sell orders and getting their position more and more inventories.  

- On 2018-01-10, if we look at the market price, we would easily find that the market is on a clearly up trend. Why can't the Trend-Follower recognize this trend? Actually, when plot out the market bid/ask price and our quote price, we find that the Trend-Follower absolutely recognize the trend, and it quote as we expected, send out the bid price of market bid price plus 20% of the spread, and the ask price of market ask price minus 20% of the spread. The problem is that at this moment, there is no sell orders in our dataset coming in, while only buy orders are coming in. Thus, the Trend -Follower did not catch any sell orders. Anothrer trouble is that at this moment, the market is intense. The ticks at this time are much more frequent than the time on the first two days when the orders are coming in. When Trend-Follower sends out its quote and the quote are not in effect (10 milliseconds later than the market), the market price has change. It gives a higher ask price in order to retreat from the ask market, but when its ask price is in effect, the market ask price has changed to a higher level than its quote ask price. Thus, it is forced to take in a lot of buy orders and get its position negative.

```python
def insert_order(count,df_orders, i, succession_time, realized_PnL, temp_PnL,long_list,short_list): # succession time in unit of millisecond    
    signal = df_orders.at[i,'Side']
    Bid = df_orders.at[i,'Bid']
    Ask = df_orders.at[i,'Ask']
    Spread = df_orders.at[i,'Spread']
    MM_Bid = df_orders.at[i,'MM_Bid']
    MM_Ask = df_orders.at[i,'MM_Ask']
    net_Pos = df_orders.at[i,'net_Pos']
    if(signal=='B'):
        if(net_Pos <= -K):
            short_list.append(temp_1.at[i,'MM_Ask'])
            # Hedge with the market (with k=0.5*b/o spread as hedge cost)
            realized_PnL += short_list[0]-df_orders.at[i,'Ask']-0.5*(df_orders.at[i,'Ask']-df_orders.at[i,'Bid'])
            executed = short_list.pop(0)
            temp_PnL = sum([x-df_orders.at[i,'Ask'] for x in short_list])
            net_Pos+=1
        elif(net_Pos<=0 and net_Pos>-K):
            short_list.append(df_orders.at[i,'MM_Ask'])
            temp_PnL = sum([x-df_orders.at[i,'Ask'] for x in short_list])
        elif(net_Pos>0):
            realized_PnL += df_orders.at[i,'MM_Ask']-long_list[0]
            executed = long_list.pop(0)
        net_Pos-=1
        
    elif(signal=='S'):
        if(net_Pos>=K):
            long_list.append(df_orders.at[i,'MM_Bid'])
            # Hedge with the market (with k=0.5*b/o spread as hedge cost)
            realized_PnL += df_orders.at[i,'Bid']-long_list[0]-0.5*(df_orders.at[i,'Ask']-df_orders.at[i,'Bid'])
            executed = long_list.pop(0)
            temp_PnL = sum([df_orders.at[i,'Bid']-x for x in long_list])
            net_Pos-=1
        elif(net_Pos>=0 and net_Pos<K):
            long_list.append(df_orders.at[i,'MM_Bid'])
            temp_PnL = sum([df_orders.at[i,'Bid']-x for x in long_list])
        elif(net_Pos<0):
            realized_PnL += short_list[0]-df_orders.at[i,'MM_Bid']
            executed = short_list.pop(0)
        net_Pos+=1
    MM_PnL = realized_PnL+temp_PnL
    current_time = df_orders.DateTime[i]
    time_delta = datetime.timedelta(milliseconds=succession_time)
    target_time = current_time + time_delta
    
    above = df_orders[:i+1+count]
    below = df_orders[i+1+count:]
    
    inserted = pd.DataFrame([(target_time, signal, 1,2,Bid,Ask,Spread,MM_Bid,MM_Ask,1,net_Pos,MM_PnL,1)])
    inserted.columns = ['DateTime','Side','Follow-up','Action','Bid','Ask','Spread','MM_Bid','MM_Ask','Order_Capture','net_Pos','MM_PnL','Beat_Market']
    df_orders = pd.concat([above, inserted, below], axis=0, sort=False)
    #df_orders.reset_index(drop=True, inplace=True)
    
    return df_orders,realized_PnL, temp_PnL, long_list,short_list
```

```python
def pos_moniter(temp_1,i,K,net_Pos,long_list,short_list,realized_PnL,temp_PnL):
    if(temp_1.at[i,'Side']=='S'):
        # Exceed the net position limit
        if(net_Pos>=K):
            long_list.append(temp_1.at[i,'MM_Bid'])
            # Hedge with the market (with k=0.5*b/o spread as hedge cost)
            realized_PnL += temp_1.at[i,'Bid']-long_list[0]-0.5*(temp_1.at[i,'Ask']-temp_1.at[i,'Bid'])
            executed = long_list.pop(0)
            temp_PnL = sum([temp_1.at[i,'Bid']-x for x in long_list])
            net_Pos-=1
        elif(net_Pos>=0 and net_Pos<K):
            long_list.append(temp_1.at[i,'MM_Bid'])
            temp_PnL = sum([temp_1.at[i,'Bid']-x for x in long_list])
        elif(net_Pos<0):
            realized_PnL += short_list[0]-temp_1.at[i,'MM_Bid']
            executed = short_list.pop(0)
        net_Pos+=1
    elif(temp_1.at[i,'Side']=='B'):
        # Exceed the net position limit
        if(net_Pos <= -K):
            short_list.append(temp_1.at[i,'MM_Ask'])
            # Hedge with the market (with k=0.5*b/o spread as hedge cost)
            realized_PnL += short_list[0]-temp_1.at[i,'Ask']-0.5*(temp_1.at[i,'Ask']-temp_1.at[i,'Bid'])
            executed = short_list.pop(0)
            temp_PnL = sum([x-temp_1.at[i,'Ask'] for x in short_list])
            net_Pos+=1
        elif(net_Pos<=0 and net_Pos>-K):
            short_list.append(temp_1.at[i,'MM_Ask'])
            temp_PnL = sum([x-temp_1.at[i,'Ask'] for x in short_list])
        elif(net_Pos>0):
            realized_PnL += temp_1.at[i,'MM_Ask']-long_list[0]
            executed = long_list.pop(0)
        net_Pos-=1
    
    return net_Pos,long_list,short_list,realized_PnL,temp_PnL
```

```python
def MM_Simulation(temp,p,q,K,X,flag):
    temp_1=temp.copy(deep=True)
    count=0
    net_Pos = 0
    temp_PnL, realized_PnL = 0, 0
    long_list, short_list = [],[]
    temp_1['Bid'].fillna(method='ffill',inplace=True)
    temp_1['Ask'].fillna(method='ffill',inplace=True)
    temp_1['Spread'].fillna(method='ffill',inplace=True)
    temp_1['Bid_MA_20'].fillna(method='ffill',inplace=True)
    temp_1['Bid_MA_5'].fillna(method='ffill',inplace=True)
    temp_1['Ask_MA_20'].fillna(method='ffill',inplace=True)
    temp_1['Ask_MA_5'].fillna(method='ffill',inplace=True)
    temp_1.at[0,'MM_Bid'], temp_1.at[0,'MM_Ask'] = temp_1.at[0,'Bid'], temp_1.at[0,'Ask']
    temp_1.at[0,'MM_PnL'] = 0
    temp_1.at[0,'net_Pos'] = 0
    temp_1.at[0,'Follow-up'] = 0
    
    for i in range(1,len(temp_1)):
        #Allocate the strategies.
        temp_1=market_strat(temp_1,i,X,flag,net_Pos)
        temp_1.at[i,'Follow-up'] = 0    
        if(temp_1.at[i,'Side']=='S'):
            # Market price is better
            if(temp_1.at[i,'Bid']>temp_1.at[i,'MM_Bid']):
                temp_1.at[i,'Order_Capture']=0
                temp_1.at[i,'net_Pos'] = net_Pos
                temp_1.at[i,'MM_PnL'] = temp_PnL+realized_PnL
            # Our price is the same as the market
            elif(temp_1.at[i,'Bid']==temp_1.at[i,'MM_Bid']):
                dice=random.random()
                # Order is randomly given to the market
                if(dice>p):
                    temp_1.at[i,'Order_Capture']=0
                # Order is randomly given to the us
                elif(dice<=p):
                    temp_1.at[i,'Order_Capture']=1
                    temp_1.at[i,'Beat_Market']=0
                    # Check out net position
                    net_Pos,long_list,short_list,realized_PnL,temp_PnL = pos_moniter(temp_1,i,K,net_Pos,long_list,short_list,realized_PnL,temp_PnL)
                temp_1.at[i,'net_Pos'] = net_Pos
                temp_1.at[i,'MM_PnL'] = temp_PnL+realized_PnL
            # Our price is better
            elif(temp_1.at[i,'Bid']<temp_1.at[i,'MM_Bid']):
                temp_1.at[i,'Order_Capture']=1
                temp_1.at[i,'Beat_Market']=1
                # Check out net position
                net_Pos,long_list,short_list,realized_PnL,temp_PnL = pos_moniter(temp_1,i,K,net_Pos,long_list,short_list,realized_PnL,temp_PnL)
                temp_1.at[i,'net_Pos'] = net_Pos
                trick = temp_PnL+realized_PnL
                temp_1.at[i,'MM_PnL'] = temp_PnL+realized_PnL
                # A follow-up order is randomly generated
                dice=random.random()
                if(dice<q and temp_1.at[i-1,'Follow-up']==0):
                    temp_1, realized_PnL, temp_PnL,long_list,short_list = insert_order(count,temp_1,i,Y,realized_PnL, temp_PnL,long_list,short_list)
                    net_Pos+=1
                    temp_1.at[i,'MM_PnL'] = trick
                    count+=1

        elif(temp_1.at[i,'Side']=='B'):
            # Market price is better
            if(temp_1.at[i,'Ask']<temp_1.at[i,'MM_Ask']):
                temp_1.at[i,'Order_Capture']=0
                temp_1.at[i,'net_Pos'] = net_Pos
                temp_1.at[i,'MM_PnL'] = temp_PnL+realized_PnL
            # Our price is the same as the market
            elif(temp_1.at[i,'Ask']==temp_1.at[i,'MM_Ask']):
                dice=random.random()
                # Order is randomly given to the market
                if(dice>p):
                    temp_1.at[i,'Order_Capture']=0
                # Order is randomly given to the us
                elif(dice<=p):
                    temp_1.at[i,'Order_Capture']=1
                    temp_1.at[i,'Beat_Market']=0
                    # Check out net position
                    net_Pos,long_list,short_list,realized_PnL,temp_PnL = pos_moniter(temp_1,i,K,net_Pos,long_list,short_list,realized_PnL,temp_PnL)
                temp_1.at[i,'net_Pos'] = net_Pos
                temp_1.at[i,'MM_PnL'] = temp_PnL+realized_PnL
            # Our price is better
            elif(temp_1.at[i,'Ask']>temp_1.at[i,'MM_Ask']):
                temp_1.at[i,'Order_Capture']=1
                temp_1.at[i,'Beat_Market']=1
                # Check out net position
                net_Pos,long_list,short_list,realized_PnL,temp_PnL = pos_moniter(temp_1,i,K,net_Pos,long_list,short_list,realized_PnL,temp_PnL)
                temp_1.at[i,'net_Pos'] = net_Pos
                trick = temp_PnL+realized_PnL
                temp_1.at[i,'MM_PnL'] = temp_PnL+realized_PnL
                # A follow-up order is randomly generated
                dice=random.random()
                
                if(dice<q and temp_1.at[i-1,'Follow-up']==0):
                    temp_1, realized_PnL, temp_PnL,long_list,short_list = insert_order(count,temp_1,i,Y,realized_PnL, temp_PnL,long_list,short_list)
                    net_Pos-=1
                    temp_1.at[i,'MM_PnL'] = trick
                    count+=1

        elif(temp_1.at[i,'Side']=='F'):
            if(net_Pos<0):
                temp_PnL = sum([x-temp_1.at[i,'Ask'] for x in short_list])
            elif(net_Pos>0):
                temp_PnL = sum([temp_1.at[i,'Bid']-x for x in long_list])
            elif(net_Pos==0):
                temp_PnL = 0
            temp_1.at[i,'net_Pos'] = net_Pos
            temp_1.at[i,'MM_PnL'] = temp_PnL+realized_PnL
        #temp_1.reset_index(drop=True, inplace=True)
    return temp_1
```

```python
overall = pd.DataFrame(columns=['Median PnL', 'Min PnL', 'Median Net Position', 'Catch or Not', 'Better than Market'], 
                       index=['Strategy 1: 2018-01-07',
                              'Strategy 1: 2018-01-08',
                              'Strategy 1: 2018-01-10',
                              'Strategy 1: 2018-01-12',
                              'Strategy 2: 2018-01-07',
                              'Strategy 2: 2018-01-08',
                              'Strategy 2: 2018-01-10',
                              'Strategy 2: 2018-01-12',
                              'Strategy 3: 2018-01-07',
                              'Strategy 3: 2018-01-08',
                              'Strategy 3: 2018-01-10',
                              'Strategy 3: 2018-01-12'])
```

```python
def Calculate_Statistics(result):
    
    s = result[result['Side']!='F'].index[0]
    e = result[result['Side']!='F'].index[-1]
    
    median_PnL = result.loc[s:e, 'MM_PnL'].median()*10000
    min_PnL = result.loc[s:e, 'MM_PnL'].min()*10000
    
    median_net_pos = result.loc[s:e, 'net_Pos'].median()
    
    Take = len(result[result['Order_Capture']==1])
    Slip = len(result[result['Order_Capture']==0])
    Catch = '{}:{}'.format(Take, Slip)
    
    yes = len(result[result['Beat_Market']==1])
    no = len(result[result['Beat_Market']==0])
    Better = '{}:{}'.format(yes, no)
    
    return median_PnL, min_PnL, median_net_pos, Catch, Better
```

```python
overall.loc['Strategy 1: 2018-01-07'] = Calculate_Statistics(result_11)
overall.loc['Strategy 1: 2018-01-08'] = Calculate_Statistics(result_21)
overall.loc['Strategy 1: 2018-01-10'] = Calculate_Statistics(result_31)
overall.loc['Strategy 1: 2018-01-12'] = Calculate_Statistics(result_41)

overall.loc['Strategy 2: 2018-01-07'] = Calculate_Statistics(result_12)
overall.loc['Strategy 2: 2018-01-08'] = Calculate_Statistics(result_22)
overall.loc['Strategy 2: 2018-01-10'] = Calculate_Statistics(result_32)
overall.loc['Strategy 2: 2018-01-12'] = Calculate_Statistics(result_42)

overall.loc['Strategy 3: 2018-01-07'] = Calculate_Statistics(result_13)
overall.loc['Strategy 3: 2018-01-08'] = Calculate_Statistics(result_23)
overall.loc['Strategy 3: 2018-01-10'] = Calculate_Statistics(result_33)
overall.loc['Strategy 3: 2018-01-12'] = Calculate_Statistics(result_43)
```
