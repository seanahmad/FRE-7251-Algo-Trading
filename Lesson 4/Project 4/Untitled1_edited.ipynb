# Introduction

In this project, we are going to run a program to act as a market maker trading GBP/USD. Market tick data and order data are given. There are 4 sets of tick data, in 4 different days respectively. And also there are 4 sets of order data, in 4 different days respectively. We are going to use these dataset to test our trading algorithm, and markdown the change of our PnL and net position along time series. At the last, we will do an analysis on the performance of our algorithm and discuss potential improvement.

While simulating trading, we strictly follow the trading rules below:
- The reaction time after every new market event is $X\ =\ 10\ milliseconds$
- Execution probability at market price is $P\ =\ 0.5$
- Follow-up order probability at better-than-market price is $Q = 0.25$
- Net open position limit is $K\ =\ 10 \ Million \ GBP$
- Maximum PnL loss limit is $L\ =\ 5000 \ Dollars = 0.005 \ Million \ Dollars$

# Strategies

We build two different strategies to do the trading simulation.
- __Market-Follower:__  
    In this strategy, we quote exactly the same as the market price observed, but we would have to be X milliseconds later than the market. For instance, when we observed the market bid price is 1.3554 and market ask price is 1.3562, we would send out the same bid and ask price, but our price is delayed X milliseconds. In the meantime, because our price is the same as the market's, we may not be able to execute the order (probability is 0.5). As is stated, this strategy is a passive management strategy. We do not intervene with how the price is quoted.


- __Position-Oriented:__  
    In this strategy, we would adjust our price quote according to our net position and the market price. We want to achieve a relatively lower net position in the whole time window, by actively manage our position with price quote. We would discuss different situation and quote below.
    1. When our net open position is between -3 and 3, the net open position is low, and we aggressively want to take more orders. Thus, we would give out a 0.0001 (1 pip) lower bid price than market bid price and a 0.0001 (1 pip) higher ask price than the market ask price.
    2. When our net open position is between -7 and -3 or between 3 and 7, the net open position is in a medium level, and we would stand neutrally in the market, i.e. follow the market price to give out our price. In this situation, we act the same as the first strategy.
    3. When our net open position is less than -7, we have taken a lot of buy orders. We would stay in the bid price market, by increasing 0.0001 (1 pip) than the market bid price, in order to take more sell orders to offset our negative positon. In the meantime, we would stay out of the ask market, by increasing 0.001 (1 pip) than the market ask price, in order to reduce the chance of taking more buy orders.
    4. When our net open position is less than -7, we have taken a lot of sell orders. We would stay in the ask price market, by lowering 0.0001 (1 pip) than the market ask price, in order to take more buy orders to offset our negative positon. In the meantime, we would stay out of the bid market, by lowering 0.001 (1 pip) than the market bid price, in order to reduce the chance of taking more sell orders.

```python
NOTEBOOK_ROOT = os.getcwd()
# Gbp/Usd Bid and Ask price in the market
M = pd.read_csv(os.path.join(NOTEBOOK_ROOT, "2.csv"), encoding='utf-16-le', index_col=0)
M_raw = pd.read_csv(os.path.join(NOTEBOOK_ROOT, "2.csv"), encoding='utf-16-le') # Without reindex
M.index = pd.to_datetime(M.index, format="%m/%d/%Y %H:%M:%S.%f")
M_raw.DateTime = pd.to_datetime(M_raw.DateTime, format="%m/%d/%Y %H:%M:%S.%f")
M['Action'] = 1
M_raw['Action']=1
M_raw['Side']='F'
M_raw['Spread']=M_raw['Ask']-M_raw['Bid']
```

```python
# Order Information
order = pd.read_csv(os.path.join(NOTEBOOK_ROOT, "MMOrderSet.csv"),index_col = False)
order['DateTime'] = pd.to_datetime(order['Time'], format = "%m-%d-%Y %H:%M:%S")
order.drop(['Time'],axis=1,inplace=True)
order = order[['DateTime', 'Side']]
order['Action'] = 2
```

```python
# Interweave orders and tick prices based on their respective times of arrival
data = pd.concat([M_raw, order], axis=0, sort=False).sort_values(by=['DateTime'])
data.reset_index(drop=True, inplace=True)
```

```python
for i in range(len(temp_1)):
    temp_1.at[i,'net_Pos'] = net_Pos
    if(temp_1.at[i,'Side']=='S'):
        # Market price is better
        if(temp_1.at[i,'Bid']>temp_1.at[i,'MM_Bid']):
            temp_1.at[i,'Order_Capture']=0
        # Our price is the same as the market
        elif(temp_1.at[i,'Bid']==temp_1.at[i,'MM_Bid']):
            p=random.random()
            # Order is randomly given to the market
            if(p>0.5):
                temp_1.at[i,'Order_Capture']=0
            # Order is randomly given to the us
            elif(p<=0.5):
                temp_1.at[i,'Order_Capture']=1
                # Exceed the net position limit
                if(net_Pos==K):
                    long_list.append(temp_1.at[i,'MM_Bid'])
                    # Hedge with the market (with k=0.5*b/o spread as hedge cost)
                    realized_PnL += temp_1.at[i,'Bid']-long_list[0]-0.5*(temp_1.at[i,'Ask']-temp_1.at[i,'Bid'])
                    executed = long_list.pop(0)
                    temp_PnL = sum([temp_1.at[i,'Bid']-x for x in long_list])
                    net_Pos-=1
                elif(net_Pos>=0 and net_Pos<K):
                    long_list.append(temp_1.at[i,'MM_Bid'])
                    temp_PnL = sum([temp_1.at[i,'Bid']-x for x in long_list])
                elif(net_Pos<0):
                    realized_PnL += short_list[0]-temp_1.at[i,'MM_Bid']
                    executed = short_list.pop(0)
                net_Pos+=1
        # Our price is better
        elif(temp_1.at[i,'Bid']<temp_1.at[i,'MM_Bid']):
            temp_1.at[i,'Order_Capture']=1
            if(net_Pos==K):
                long_list.append(temp_1.at[i,'MM_Bid'])
                # Hedge with the market (with k=0.5*b/o spread as hedge cost)
                realized_PnL += temp_1.at[i,'Bid']-long_list[0]-0.5*(temp_1.at[i,'Ask']-temp_1.at[i,'Bid'])
                executed = long_list.pop(0)
                temp_PnL = sum([temp_1.at[i,'Bid']-x for x in long_list])
                net_Pos-=1
            elif(net_Pos>=0 and net_Pos<K):
                long_list.append(temp_1.at[i,'MM_Bid'])
                temp_PnL = sum([temp_1.at[i,'Bid']-x for x in long_list])
            elif(net_Pos<0):
                realized_PnL += short_list[0]-temp_1.at[i,'MM_Bid']
                executed = short_list.pop(0)
            net_Pos+=1
            # A follow-up order is randomly generated
            q=random.random()
            
            
            
    elif(temp_1.at[i,'Side']=='B'):
        # Market price is better
        if(temp_1.at[i,'Ask']<temp_1.at[i,'MM_Ask']):
            temp_1.at[i,'Order_Capture']=0
        # Our price is the same as the market
        elif(temp_1.at[i,'Ask']==temp_1.at[i,'MM_Ask']):
            p=random.random()
            # Order is randomly given to the market
            if(p>0.5):
                temp_1.at[i,'Order_Capture']=0
            # Order is randomly given to the us
            elif(p<=0.5):
                temp_1.at[i,'Order_Capture']=1
                # Exceed the net position limit
                if(net_Pos == -K):
                    short_list.append(temp_1.at[i,'MM_Ask'])
                    # Hedge with the market (with k=0.5*b/o spread as hedge cost)
                    realized_PnL += short_list[0]-temp_1.at[i,'Ask']-0.5*(temp_1.at[i,'Ask']-temp_1.at[i,'Bid'])
                    executed = short_list.pop(0)
                    temp_PnL = sum([x-temp_1.at[i,'Ask'] for x in short_list])
                    net_Pos+=1
                elif(net_Pos<=0 and net_Pos>-K):
                    short_list.append(temp_1.at[i,'MM_Ask'])
                    temp_PnL = sum([x-temp_1.at[i,'Ask'] for x in short_list])
                elif(net_Pos>0):
                    realized_PnL += temp_1.at[i,'MM_Ask']-long_list[0]
                    executed = long_list.pop(0)
                net_Pos-=1
        # Our price is better
        elif(temp_1.at[i,'Ask']>temp_1.at[i,'MM_Ask']):
            temp_1.at[i,'Order_Capture']=1
            # Exceed the net position limit
            if(net_Pos == -K):
                short_list.append(temp_1.at[i,'MM_Ask'])
                # Hedge with the market (with k=0.5*b/o spread as hedge cost)
                realized_PnL += short_list[0]-temp_1.at[i,'Ask']-0.5*(temp_1.at[i,'Ask']-temp_1.at[i,'Bid'])
                executed = short_list.pop(0)
                temp_PnL = sum([x-temp_1.at[i,'Ask'] for x in short_list])
                net_Pos+=1
            elif(net_Pos<=0 and net_Pos>-K):
                short_list.append(temp_1.at[i,'MM_Ask'])
                temp_PnL = sum([x-temp_1.at[i,'Ask'] for x in short_list])
            elif(net_Pos>0):
                realized_PnL += temp_1.at[i,'MM_Ask']-long_list[0]
                executed = long_list.pop(0)
            net_Pos-=1
            # A follow-up order is randomly generated
            q=random.random()
            
            
            
            
    elif(temp_1.at[i,'Side']=='F'):
        if(net_Pos<0):
            temp_PnL = sum([x-temp_1.at[i,'Ask'] for x in short_list])
        elif(net_Pos>0):
            temp_PnL = sum([temp_1.at[i,'Bid']-x for x in long_list])
        elif(net_Pos==0):
            temp_PnL = 0 
    temp_1.at[i,'MM_PnL'] = temp_PnL+realized_PnL
```
